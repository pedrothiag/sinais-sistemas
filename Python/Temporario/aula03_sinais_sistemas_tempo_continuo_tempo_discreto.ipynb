<h3> Gerando Sinais de Tempo Discreto </h3>

Ao compreendermos como gerar e manipular vetores em Python, torna-se possível gerar sinais de tempo discreto. Nesse contexto, é comum a criação de dois vetores distintos: o vetor `n`, que representa o tempo discreto e o vetor `x`, que contém os valores do sinal em cada instante de tempo.

A plotagem de sinais de tempo discreto pode ser facilmente realizada em Python através da biblioteca `matplotlib.pyplot`. Para plotar um sinal de tempo discreto. Podemos definir os rótulos dos eixos `x` e `y` utilizando as funções `xlabel` e `ylabel`, respectivamente. Além disso, é possível usar a função `title` para definir o título do gráfico. É importante ressaltar que, para visualizar o gráfico plotado, é necessário utilizar a função `show`. Essa função é responsável por exibir o gráfico plotado na janela de exibição. 

<b> Exemplo 3.1: </b> Vamos gerar um sinal de tempo discreto.

```python
import numpy as np
import matplotlib.pyplot as plt

n = [-2,-1,0,1,2,3,4]
x = [2,3,-1,-2,1.5,0,-2.5]

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.show()
```

Uma alternativa para gerar um sinal de tempo discreto em Python é utilizar vetores `n` e `x`. O vetor `n` representa a variável independente, variando de um limite inferior `n_min` a um limite superior `n_max` com incrementos de 1. Já o vetor `x` é calculado em função do vetor `n` e representa a amplitude do sinal em cada instante de tempo. 

<b> Exemplo 3.2: </b> Vamos gerar o sinal de tempo discreto dado por $x[n] = \cos (n\pi /4)$ para $0 \leq n \leq 30$.

```python
n = np.arange(0,31)
x = np.cos(n*np.pi/4)

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.show()
```

Uma funcionalidade importante na plotagem de gráficos em Python é a capacidade de exibir vários plots simultaneamente, o que pode ser feito através da função `subplot`.Com essa função, é possível criar uma matriz de plots na mesma figura e exibir diferentes gráficos lado a lado ou um abaixo do outro, permitindo a visualização de múltiplos resultados ou comparações em um único plot.

A função `subplot` aceita três argumentos:
<ul>
<li> número de linhas na matriz; </li>
<li> número de colunas; </li>
<li> índice do plot atual. </li>
</ul>
Esses argumentos indicam a posição do plot atual na matriz, permitindo que diferentes gráficos sejam exibidos em cada posição da matriz.

Alguns sinais, como o impulso e o degrau podem ser gerados através de atribuições condicionais, que assumiram valor igual à `1` quando a condição for verdadeira.

<b> Exemplo 3.3: </b> Vamos gerar o impulso unitário $h[n]$ e o degrau unitário $u[n]$. Relemebrando que:
$$
h[n] = \begin{cases}
1 & n = 0 \\
0 & \mathrm{c.c.}
\end{cases}
$$
$$
u[n] = \begin{cases}
1 & n \geq 0 \\
0 & \mathrm{c.c.}
\end{cases}
$$

```python
n = np.arange(-10,20)
delta = 1*(n==0);
step = 1*(n>=0);

plt.subplot(211)
plt.stem(n,delta)
plt.xlabel('n')
plt.ylabel('$\delta[n]$')

plt.subplot(212)
plt.stem(n,step)
plt.xlabel('n')
plt.ylabel('u[n]')

plt.tight_layout()
plt.show()
```

<b> Exemplo 3.4: </b> Vamos gerar diferentes sinais exponenciais reais, ou seja:
$$
x[n] = \alpha^n
$$
considerando: (a) $\alpha = 0,8$, (b) $\alpha = 1,2$, (c) $\alpha = -0,8$ e (d) $\alpha = -1,2$. Vamos verificar o comportamento de crescimento ou decaimento para cada caso.

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0,20)
x1 = 0.8**n;
x2 = 1.2**n;
x3 = (-0.8)**n;
x4 = (-1.2)**n;

plt.subplot(221)
plt.stem(n,x1)
plt.xlabel('n')
plt.title('$\\alpha = 0,8$')

plt.subplot(222)
plt.stem(n,x2)
plt.xlabel('n')
plt.title('$\\alpha = 1,2$')

plt.subplot(223)
plt.stem(n,x3)
plt.xlabel('n')
plt.title('$\\alpha = -0,8$')

plt.subplot(224)
plt.stem(n,x4)
plt.xlabel('n')
plt.title('$\\alpha = -1,2$')

plt.tight_layout()
plt.show()
```

Se um determinado sinal é complexo, temos que plotar a parte real (usando `np.real`), parte imaginária (usando `np.imag`), magnitude (usando `np.abs`) ou fase (usando `np.fase`).

<b> Exemplo 3.5: </b> Vamos gerar diferentes sinais exponenciais complexos, ou seja:
$$
x[n] = C\alpha^n
$$
considerando $C = 1,2e^{j\pi /10}$ e: (a) $\alpha = 1,2e^{j\pi /4}$, (b) $\alpha = 0,8e^{j\pi /4}$. Vamos verificar o comportamento de crescimento ou decaimento para parte real em cada caso.

```python
n = np.arange(0,20)

C = 1.2*np.exp(1j*np.pi/10)
alpha1 = 1.2*np.exp(1j*np.pi/4);
x1 = C*alpha1**n;

alpha2 = 0.8*np.exp(1j*np.pi/4);
x2 = C*alpha2**n;

plt.subplot(211)
plt.stem(n,np.real(x1))
plt.xlabel('n')
plt.title('$|\\alpha| > 1$')

plt.subplot(212)
plt.stem(n,np.real(x2))
plt.xlabel('n')
plt.title('$|\\alpha| < 1$')

plt.tight_layout()
plt.show()
```

<h3> Gerando Sinais de Tempo Contínuo </h3>

Agora que já aprendemos a gerar sinais de tempo discreto, vamos aprender a gerar sinais de tempo contínuo. Para isso, utilizaremos dois vetores: `t` e `x`. O vetor `t` representará a variável independente e variará entre um limite inferior `t_min` e um limite superior `t_max` com um passo de `T_s`. Já o vetor `x` será calculado em função do vetor `t` e representará a amplitude do sinal.

Na prática, gerar um sinal de tempo contínuo computacionalmente significa calcular a expressão do sinal para diferentes valores de `t`, sendo que esses valores são tão próximos um do outro que visualmente não notamos diferença do sinal contínuo para esses valores calculados em alguns pontos. 

Para determinar o valor adequado de `T_s`, é preciso utilizar o Teorema da Amostragem. No entanto, para fins de aprendizado, usaremos valores pequenos, geralmente na ordem de $10^{-4}$. 

Uma vez que o vetor `t` e o vetor `x` tenham sido gerados, podemos plotar o sinal contínuo utilizando a função `plot` da biblioteca `matplotlib.pyplot`. Podemos definir os rótulos dos eixos `x` e `y` utilizando as funções `xlabel` e `ylabel`, respectivamente. Além disso, é possível usar a função `title` para definir o título do gráfico.

É importante lembrar que, assim como no caso dos sinais de tempo discreto, é possível utilizar a função `subplot` para plotar vários sinais de tempo contínuo em um único gráfico.

<b> Exemplo 3.6: </b> Vamos gerar o sinal de tempo contínuo dado por $x(t) = \cos (2 \pi 10 t)$ para $0 < t < 1$. Utilizaremos um passo de $T = 10^{-4}$ s.

```python
t = np.arange(0,1,1e-4)
x = np.cos(2*np.pi*10*t)

plt.plot(t,x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.show()
```

<b> Exemplo 3.7: </b> Vamos plotar o degrau unitário de tempo contínuo. Relembrando:
$$
u(t) = \begin{cases}
1 & t \geq 0 \\
0 & \mathrm{c.c.}
\end{cases}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-2,5,1e-4)
x = 1*(t>=0);

plt.plot(t,x)
plt.xlabel('t')
plt.ylabel('u(t)')
plt.show()
```

O impulso unitário de tempo contínuo é uma função matemática que apresenta uma amplitude infinita em um instante de tempo específico e zero em todos os outros instantes. Essa característica de ter amplitude infinita em um ponto específico faz com que o impulso não possa ser gerado computacionalmente, pois não há como representar numericamente um número infinito em um ponto específico. Todavia, esse sinal pode ser gerado de forma simbólica. Aprenderemos um pouco mais sobre computação simbólica mais à frente

<b> Exemplo 3.8: </b> Dado os seguintes sinais:
$$
x_1[n] = \sin (\pi/4 n)
$$
$$
x_2[n] = u[n]
$$
$$
x_3[n] = 0,25n
$$
Determine os seguintes sinais: (a) $x_1[n] + x_2[n]$, (b) $2x_1[n]$, (c) $x_1[n]x_2[n]$ e (d) $1,5x_1[n] + x_2[n] + 2,5x_3[n]$

```python
n = np.arange(-10,20)
x1 = np.sin(np.pi/4*n)
x2 = (n>=0)
x3 = 0.25*n

xadd = x1 + x3
xgain = 2*x1
xmul = x1*x2
xcomb = 1.5*x1 + x2 + 2.5*x3

plt.figure()
plt.subplot(311)
plt.stem(n,x1)
plt.xlabel('n')
plt.title('$x_1[n]$')

plt.subplot(312)
plt.stem(n,x2)
plt.xlabel('n')
plt.title('$x_2[n]$')

plt.subplot(313)
plt.stem(n,x3)
plt.xlabel('n')
plt.title('$x_3[n]$')
plt.tight_layout()

plt.figure()
plt.subplot(221)
plt.stem(n,xadd)
plt.xlabel('n')
plt.title('$x_1 + x_2[n]$')

plt.subplot(222)
plt.stem(n,xgain)
plt.xlabel('n')
plt.title('$2x_1[n]$')

plt.subplot(223)
plt.stem(n,xmul)
plt.xlabel('n')
plt.title('$x_1[n]x_3[n]$')

plt.subplot(224)
plt.stem(n,xcomb)
plt.xlabel('n')
plt.title('$1.5x_1[n] + x_2[n] + 0.5x_3[n]$')
plt.tight_layout()

plt.show()
```

<b> Exemplo 3.9: </b> Dado o seguinte sinal:

$$
x(t) = \begin{cases}
t + 1 & -1 \leq t \leq 0 \\
-0,5t + 1 & 0 < t \leq 2 \\
0 & \mathrm{c.c.}
\end{cases}
$$

Determine: (a) $x(t/2)$, (b) $x(2t)$, (c) $x(-t)$.

```python
def meusinal(t):
    x = (t+1)*(np.logical_and(t>=-1,t<=0)) + (-0.5*t+1)*(np.logical_and(t>0,t<=2))
    return x

t = np.arange(-4,4,1e-4)
x = meusinal(t)

x1 = meusinal(t/2)
x2 = meusinal(2*t)
x3 = meusinal(-t)

plt.figure()
plt.subplot(411)
plt.plot(t,x)
plt.xlabel('t')
plt.title('$x(t)$')

plt.subplot(412)
plt.plot(t,x1)
plt.xlabel('t')
plt.title('$x(t/2)$')

plt.subplot(413)
plt.plot(t,x2)
plt.xlabel('t')
plt.title('$x(2t)$')

plt.subplot(414)
plt.plot(t,x3)
plt.xlabel('t')
plt.title('$x(-t)$')

plt.tight_layout()
plt.show()
```

<b> Exemplo 3.10: </b> Vamos simular um sistema de tempo discreto correspondente a uma média móvel de três pontos:
$$
y[n] = \frac{1}{3}(x[n] + x[n-1] + x[n-2])
$$
Considerando que a entrada é:
$$
x[n] = \sin (\pi/16n)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0,60)
x = np.sin(np.pi/16*n) + 0.15*np.random.randn(len(n))

y = np.zeros(len(n))
y[0] = 1/3*x[0]
y[1] = 1/3*(x[1] + x[0])
for index in np.arange(2,len(n)):
    y[index] = 1/3*(x[index] + x[index-1] + x[index-2])

plt.subplot(211)
plt.stem(n,x)
plt.xlabel('n')
plt.title('$x[n]$')

plt.subplot(212)
plt.stem(n,y)
plt.xlabel('n')
plt.title('$y[n]$')

plt.tight_layout()
plt.show()
```
