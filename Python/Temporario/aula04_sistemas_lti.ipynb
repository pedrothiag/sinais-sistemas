```python
import numpy as np
import matplotlib.pyplot as plt

dt = 1e-4
t_min = -2;
t_max = 4;
t = np.arange(t_min,t_max,dt)

T = 1;
x = (0 < t) & (t < T);
h = ((0 < t) & (t < 2*T))*t;

plt.subplot(211)
plt.plot(t,x)
plt.title('x(t)')
plt.xlabel('t')
plt.subplot(212)
plt.plot(t,h)
plt.title('h(t)')
plt.xlabel('t')
plt.tight_layout()
plt.show()
```

```python
y = dt*np.convolve(x,h)
ty = np.arange(2*t_min,2*t_max-dt,dt)

plt.plot(ty,y)
plt.title('y(t) = x(t)*h(t)')
plt.xlabel('t')
plt.show()
```

```python
dt = 1e-4
t_min = -2;
t_max = 4;
t = np.arange(t_min,t_max,dt)

x = np.exp(-2*t)*(t>=0)
h = (t>=0)

y = dt*np.convolve(x,h)
ty = np.arange(2*t_min,2*t_max-dt,dt)

plt.subplot(311)
plt.plot(t,x)
plt.title('x(t)')
plt.xlabel('t')

plt.subplot(312)
plt.plot(t,h)
plt.title('h(t)')
plt.xlabel('t')

plt.subplot(313)
plt.plot(ty,y)
plt.title('y(t) = x(t)*h(t)')
plt.xlabel('t')
plt.tight_layout()
plt.show()
```
