<h1> Séries de Fourier </h1>

<b> Exemplo 1: </b> Vamos analisar a decomposição em série de Fourier do sinal do Exemplo 5.2 dos Slides.

```python
import numpy as np
import matplotlib.pyplot as plt

Ts = 1e-4
t_period = np.arange(0,np.pi,Ts)
x_period = np.exp(-t_period/2)

rep = 3
x = np.tile(x_period,rep)
t = np.arange(0,rep*np.pi,Ts)

plt.plot(t,x)
plt.xlim([np.min(t),np.max(t)])
plt.xlabel('$t$')
plt.title('$x(t)$')
plt.xticks([0,np.pi,2*np.pi,3*np.pi],['0','$\pi$','$2\pi$','$3\pi$'])
plt.show()
```

```python
k = np.arange(-10,11)
ak = 0.504/(1 + 4j*k)

plt.subplot(2,1,1)
plt.stem(k,np.abs(ak))
plt.xlabel('$k$')
plt.title('$|a_k|$')
plt.xticks(k)

plt.subplot(2,1,2)
plt.stem(k,np.angle(ak))
plt.xlabel('$k$')
plt.title('$|a_k|$')
plt.xticks(k)
plt.yticks([-np.pi/2,0,np.pi/2],['$\pi/2$','0','$\pi/2$'])

plt.tight_layout()
plt.show()
```

```python
Nmax = 40
k = np.arange(0,Nmax)
ak = 0.504/(1 + 4j*k)
ak_mag = np.abs(ak)
ak_phase = np.angle(ak)

x_Fourier10 = 0.504*np.ones(len(x))
for kindex in np.arange(1,5):
    x_Fourier10 = x_Fourier10 + 2*ak_mag[kindex]*np.cos(2*kindex*t + ak_phase[kindex])

x_Fourier30 = 0.504*np.ones(len(x))
for kindex in np.arange(1,15):
    x_Fourier30 = x_Fourier30 + 2*ak_mag[kindex]*np.cos(2*kindex*t + ak_phase[kindex])

plt.plot(t,x)
plt.xlim([np.min(t),np.max(t)])
plt.xlabel('$t$')
plt.xticks([0,np.pi,2*np.pi,3*np.pi],['0','$\pi$','$2\pi$','$3\pi$'])
plt.plot(t,x_Fourier10,'--')
plt.plot(t,x_Fourier30,'--')
plt.legend(['$x(t)$','$N$ = 5','$N$ = 15'])
plt.show()
```
