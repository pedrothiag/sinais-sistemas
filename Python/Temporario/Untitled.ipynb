```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = "Arial"

# Define Transfer Function
num = np.array([2])
den = np.array([3 , 1])
H = signal.TransferFunction(num , den)

# Impulse Response
t, y = signal.impulse(H)

#Freqency Response
w, Hw = signal.freqresp(H)

# Plotting
plt.plot(t, y)
plt.title("Step Response")
plt.xlabel("t")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
import numpy as np
import control as ctrl
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = "Arial"

# Define Transfer Function
num = np.array([2])
den = np.array([3, 1])
H = ctrl.tf(num, den)

t = np.linspace(0, 5, 1000)  # Vetor de tempo para a simulao
u = np.heaviside(t, 1)  # Funo degrau unitrio

x = np.exp(-2 * t) * u  # Entrada e^(-2t)u(t)

tout, yout = ctrl.forced_response(H, t, x)

plt.plot(tout, yout)
plt.xlabel('Tempo')
plt.ylabel('Sada do Sistema')
plt.title('Resposta do sistema  entrada $\exp(-2t)u(t)$')
plt.grid(True)
plt.show()
```
